```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import tikzplotlib as tkpl
```

```python
x = np.random.uniform(0,1,20)
y = x**2
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,6))
ax.plot(x,'or--',lw=3)
ax.set_title('This is a title')
ax.set_xlabel('Index $(cm^2)$')
ax.set_ylabel('Value (mm $^3$)')
tkpl.save("../images/rtexandomplot.")
fig.savefig("../images/randomplot.png",dpi=150)
```
